```python
def convert_fname(cell):
  if(cell=='Aleshia'):
    return 'Aakanksha'
  return cell
def convert_lname(cell):
  if(cell=='Andrade'):
    return 'Trivedi'
  return cell
df4=pd.read_excel('/content/Sample_Employee_data_xls.xls',converters={'first_name':convert_fname,
                                                                      'last_name':convert_lname})
df4
```

```python
df4.to_excel("Excelfile.xlsx",sheet_name="Employee",index=False,startrow=5,startcol=6)
```
