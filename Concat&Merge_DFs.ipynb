```python
import pandas as pd
india_weather={
    "city":["mumbai","delhi","Banglore"],
    "temp":[32,45,40],
    "humidity":[80,6,78]
}
df1=pd.DataFrame(india_weather)
df1
```

```python
us_weather={
    'city':["New York","Chicago","Orlando"],
    'temp':[21,24,34],
    'humidity':[68,65,76]
}
df2=pd.DataFrame(us_weather)
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df=pd.concat([df1,df2])
df
```

```python
df=pd.concat([df1,df2],keys=['India','USA'])
df
```

```python
#Merging Two dataFrame
temp=pd.DataFrame({
    "city":["mumbai","delhi","Banglore",'Pune'],
    "temp":[32,45,40,76]})
wind=pd.DataFrame({
    "city":["delhi","mumbai","Banglore","hyderabad"],
    "wind":[32,45,40,21]})
df6=pd.merge(temp,wind,on="city")
df6
```

```python
df7=pd.merge(temp,wind,how="outer")
df7
```

```python
df8=pd.merge(temp,wind,how="left")
df8
```

```python
df9=pd.merge(temp,wind,how="right")
df9
```

```python
df10=pd.merge(temp,wind,how="right",indicator=True)
df10
```

```python
india_weather={
    "city":["mumbai","delhi","Banglore"],
    "temp":[32,45,40],
    "humidity":[80,6,78]
}
dff=pd.DataFrame(india_weather)
dff
```

```python
us_weather={
    'city':["New York","Chicago","delhi"],
    'temp':[21,24,34],
    'humidity':[68,65,76]
}
dff2=pd.DataFrame(us_weather)
dff2
```

```python
dff3=pd.merge(dff,dff2,on="city",suffixes=('_left','_right'))
dff3
```
