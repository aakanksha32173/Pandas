```python
df=pd.read_csv('/content/weather_data.csv',parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index("day",inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df2=df.fillna({
    'temperature':df['temperature'].mean(),
    'windspeed':df['windspeed'].mean(),
    'event':'no event'
})
new_df2
```

```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
new_df=df.fillna(method='bfill',axis=1)
new_df
```

```python
new_df=df.fillna(method='bfill',limit=2)
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
newdf=df.dropna()
newdf
```

```python
newdf=df.dropna(how="all")
newdf
```

```python
newdf=df.dropna(thresh=2)
newdf
```

```python
dt=pd.date_range('01-01-2017','11-01-2017')
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
