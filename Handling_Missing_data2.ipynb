```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/weather_data (1).csv")
df
```

```python
newdf=df.replace(-99999,np.nan)
newdf
```

```python
newdf=df.replace([-99999,"0"],np.nan)
newdf
```

```python
newdf=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.nan)
newdf
```

```python
newdf=df.replace({
    -99999:np.nan,
    '0':'Sunny'
})
newdf
```

```python
#Replacing with regex
df2=pd.read_csv('/content/weather_data (1).csv')
df2
```

```python
newdf=df2.replace('[A-Za-z]','',regex=True)
newdf
```

```python
newdf=df2.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
newdf
```

```python
#Replace a list of values with another
df=pd.DataFrame(
    {
        'name':['surbhi','Sameeksha','Aakanksha'],
        'score':['good','bad','worse']
    }
)
df
```

```python
newdf=df.replace(['good','bad','worse'],[3,1,0])
newdf
```
