```python
import pandas as pd
#Reading a csv file
df=pd.read_csv('/content/tmdb_5000_movies.csv')
df
```

```python
#creating dataframe using a python dictionary
data={
    'name':['Aakanksha','Prachi','Pranjal','Surbhi','Kanha','Anant'],
    'Marks':[50,60,70,30,40,90]
}
df3=pd.DataFrame(data)
df3
```

```python
#creating data frame using tuples
data2={
    ('1','Chirag'),
    ('2','Anaya'),
    ('3','Shruti')
}
df4=pd.DataFrame(data2,columns=['SNo','Name'])
df4
```

```python
#creating dataframe from multiple dictionaries
data3=[
    {'Name':'Aakanksha','RNo':'20C6001'},
    {'Name':'Surbhi','RNo':'20C6054'},
    {'Name':'Sameeksha','RNo':'20C6043'}
]

df5=pd.DataFrame(data3)
df5
```
